```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.histplot(data= titanic , x='age' ) # 히스토그램은 수치형 데이터의 구간별 빈도수
```

```python
sns.histplot(data= titanic , x='age' , bins = 10 )
```

```python
sns.histplot(data= titanic , x='age' , hue = 'alive') # 범주별로 구분 데이터
```

```python
sns.histplot(data= titanic , x='age'  , hue = 'alive' , multiple= 'stack') # 생존자 누적 그래프
```

```python
sns.kdeplot(data = titanic , x ='age') # 커널밀도추정 함수 그래프(히스토그램을 매끄럽게 곡선으로 연결한 그래프)
```

```python
sns.kdeplot(data = titanic , x ='age' , hue = 'alive' , multiple ='stack') # 누적 커널밀도추정 함수 그래프
```

```python
sns.displot(data=titanic , x = 'age') # 분포도 : 수치형 데이터 하나의 분포를 나타내는 그래프
```

```python
sns.displot(data=titanic , x='age' , kind = 'kde')
```

```python
sns.displot(data=titanic , x='age' , kde = True)
```

```python
sns.kdeplot(data=titanic , x='age')
sns.rugplot(data=titanic , x='age') # 러그플롯은 주변 분포를 나타내는 그래프
```

```python
sns.barplot(x='class' , y='fare', data=titanic) # x축에 범주형 데이터, y축에 수치형 데이터
```

```python
sns.pointplot(x='class' , y='fare' , data= titanic)
```

```python
sns.boxplot(x='class' , y='age' , data=titanic)
```

```python
sns.violinplot(x='class' , y='age' , data=titanic)
#바이올린플롯은 박스플롯과 커널밀도추정 함수 그래프를 합쳐 놓은 그래프이다.
```

```python
sns.violinplot(x='class' , y='age' , hue = 'sex' ,data=titanic  , split= True) # 수치형 데이터의 전체적인 분포 양상을 알기 위해서는 바이올릿 플롯이 좋다.
```

```python
sns.countplot(x='class' , data= titanic)
```

```python
sns.countplot(y='class' , data= titanic)

# barplot은 범주형 데이터 , 수치형 데이터 2개의 피처
# countplot은 범주형 데이터 하나만 받는다.
```

```python
sns.barplot(x='class' , y='fare' , data=titanic , estimator= np.median)
```

```python
sns.barplot(x='class' , y='fare' , data=titanic , estimator= np.max)
```

```python
sns.barplot(x='class' , y='fare' , data=titanic , estimator= np.min)
```

```python
flights = sns.load_dataset('flights') # 비행기 탑승자 수 데이터 불러오기
```

```python
flights.head()
```

```python
flights_pivot = flights.pivot(index = 'month' , columns = 'year' , values = 'passengers')

flights_pivot
```

```python
sns.heatmap(data=flights_pivot)
```

```python
sns.lineplot(x='year' , y='passengers' , data=flights) # 전달한 값의 평균과 95% 신뢰구간을 나타낸다.
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.scatterplot(x='total_bill' , y ='tip' , data= tips)
```

```python
sns.scatterplot(x='total_bill' , y='tip' , hue='time' , data= tips) # 범주형 데이터별로 나누어 그릴 수 있다.
```

```python
sns.regplot(x='total_bill' , y ='tip' , data= tips) # regplot()은 선형 회귀선을 동시에 그려주는 함수이다.
```

```python
sns.regplot(x='total_bill' , y ='tip' ,ci = 99 , data= tips) # 신뢰구간 99%에 대한 regplot()은 선형 회귀선을 동시에 그려주는 함수이다.
```
