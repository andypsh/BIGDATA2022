```python
x = Symbol('x')

f_x = (x+1) /2
f_x_m = f_x * x
print(f_x_m)
mean_1 = Integral(f_x_m, (x,-1 , 1)).doit()

print("평균 : {}".format(mean_1))
f_x_v = f_x_m * x
print(f_x_v)
vars_1 = Integral(f_x_v , (x,-1,1)).doit() - mean_1**2
print("분산 : {}".format(vars_1))

ratio = Integral(f_x , (x,0,0.5)).doit()
print("확률 0이상 0.5 이하는 {}이다.".format(ratio))
```
